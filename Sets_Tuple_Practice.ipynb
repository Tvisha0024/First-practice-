```python
n_str = input("Enter the numbers separated by commas: ")
n_list = [int(x) for x in n_str.split(',')]
Tuple = tuple(n_list)
print("Tuple: ", Tuple)
Reverse = Tuple[::-1]
print("Reversed: ", Reverse)
total_sum = sum(Tuple)
print("Sum: ", total_sum)
```

```python
try:
  data= tuple(map(int, input().split()))
except:
  print("Invalid Input")
  exit()

print("Tuple:",data)
print("Reversed:",data[::-1])
print("Sum:",sum(data))
print("Average",round(sum(data/len(data)),2))
```

```python
vals= input().split()

if len(vals) !=3:
  print("Invalid Data")
else:
  name,age,city = vals
  if not age.isdigit() or name.strip()=="" or city.strip()=="":
    print("Invalid data")
  else:
    t= (name,int(age),city)
    n,a,c = t
    print(f"Name: {n}, Age: {a}, City: {c}")
```

```python
try:
    Set1_str = input("Enter elements for Set 1 (comma-separated integers): ")
    Set1 = set(map(int, Set1_str.split(',')))

    Set2_str = input("Enter elements for Set 2 (comma-separated integers): ")
    Set2 = set(map(int, Set2_str.split(',')))

    SetC = Set1.union(Set2)
    print("Union: ", SetC)

    SetD = Set1.intersection(Set2)
    print("Intersecation: ",SetD)

    SetE = Set1.difference(Set2)
    print("Difference: ",SetE)

except ValueError:
    print("Invalid input. Please enter comma-separated integers.")
```
